#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python

high_value_house = ds.loc[ds["price"].idxmax()]
print(high_value_house["price"])

address = high_value_house["address"]

print(f"La casa con dirección en {address} tiene un precio de {high_value_house['price']}.")



```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Este ejercicio es similar al ejercicio anterior, solo que ahora buscamos la casa con el precio más bajo. Recuerda imprimir la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000 USD`

```python
ds_filtered = ds[ds["price"] > 0]  
low_value_house = ds_filtered.loc[ds_filtered["price"].idxmin()]
address1 = low_value_house["address"]  
price1 = low_value_house["price"]

print(f"La casa con dirección en {address1} tiene un precio de {price1}.")





```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt

arroyomolinos_data = ds[ds["level5"] == "Arroyomolinos (Madrid)"]

plt.figure(figsize=(10, 6))
plt.hist(arroyomolinos_data["price"], bins=30, edgecolor='black', color='blue')
plt.xlabel('Precio (€)')
plt.ylabel('Frecuencia')
plt.title('Distribución de Precios en Arroyomolinos (Madrid)')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

The distributions of precios arroyomolino (madrid) is normal distribution with bell shape. wich show our mean price around 300000 euros and a range price of house between 20000 and 400000
